### Para encontrar el diametro, lo que se hizo fue lo siguiente. Primero se transformaron los datos en el archivo de datos "circulos.dat" en una matriz NxN, donde se le daba el valor de 1 ó 0 a cada posición, dependiendo si habia o no puntos en el cuadrado dlxdl correspondiente a esa entrada de la matriz (dl= 6/N) . Luego se calculo la siguiente autocorrelación:  $$ AutocorrelaciónTotal = \sum_{j=0}^{N-1} \mathscr{F}^{-1}\{|\mathscr{F}\{col_j\}|²\} $$  Esta autocorrelación es la suma de las autocorrelaciones de cada columna( no hubiera importado si se usa filas o columnas). Ahora, para extraer información de esta autocorrelación, se toma el módulo de la AutocorrelaciónTotal se analiza el resultado. Se espera que exista en el origen un maximo de autocorrelación, lo cual es un teorema probado ( http://mathworld.wolfram.com/Autocorrelation.html ).   El resultado obtenido (la ultima gráfica) se analizo de la siguiente forma. La función que se obtuvo es simetrica con respecto al desplazamiento 0. Se notan 2 máximos cercanos al 0 para los cuales se desconoce su interpretación, sin embargo se puede descartar esta distancia como el diametro por lo siguiente. La autocorrelación tiene la siguiente definición para una dimensión:  $$ Correlación(j) = \sum_{i} x_i x_{i+j} $$  Como cada columna está hecha de 1's  y 0's, se espera que cuando se desplaza cualquier distancia correspondiente a la longitud entre dos puntos tales que uno sea el reflejo del otro con respecto a el eje "x" ( un ejemplo de dos puntos asi seria como los punto $(x,y)$ y $(x,-y)$ con $(x-x_i)² + (y-y_i)² = R² $) , se obtendra un valor mayor que si se desplaza una distancia mayor al maximo de las distnacias reflejo, es decir al diametro. Este efecto no se vera casi sobre una sola correlación, por eso se suma el valor de todas las autocorrelaciones, para que se noten menos las distancias que debido al ruido aparecen una que otra vez para una sola linea pero que para otras no. Juntando esto se concluye que la distancia para la cual decae dramaticamente la autocorrelacióntotal es el diametro. Tambien se nota que para este valor la autocorrelacióntotal tiene un máximo, para el cual no se conoce la interpretación. Sin embargo utilizando la interpretación de la decaida dramatica se encuentra que el diametro de los círculos es:  $$ diametro = 1.57 $$

```python
puntos = loadtxt("./circulos.dat")
x = puntos[:,0]
y = puntos[:,1]

fig = figure(figsize=(12, 12))

scatter(x,y,s = 0.05)

xlim([-3,3])
ylim([-3,3])
```

```python
N = 1200
dl = 6.0/N

matriz = zeros((N,N))
```

```python
l_x = -3.0

for i in range(N):
    
    l_y = -3.0
    
    for j in range(N):
        
        inds = where( (x>=l_x) & (x<l_x+dl) & (y>=l_y) & (y<l_y+dl) )
        
        if( len(inds[0]) != 0):
            matriz[(i,j)] = 1.0
            
        l_y = l_y + dl

    l_x = l_x + dl
```

```python
fig = figure(figsize=(12, 12))
imshow(matriz, cmap='bone')
```

```python
autocorrelacion = zeros(N, dtype = complex)

for j in range(N):

    col = zeros(N)
    
    for i in range(N):
        
        col[i] = matriz[(i,j)]
           
    transformada_columna = ( abs( fft(col)/N ) )**2
    
    autocorrelacion = autocorrelacion + ifft(transformada_columna) 
```

```python
plot(linspace(-3,3,N-1),abs(autocorrelacion[1:]))
```

```python
autocorrelacion_positiva = autocorrelacion[N/2:]
autocorrelacion_negativa = autocorrelacion[1:N/2]

nueva_autocorrelacion = concatenate((autocorrelacion_positiva, array([autocorrelacion_positiva[0]]), autocorrelacion_negativa))
```

```python
plot(linspace(-3,3,N), abs(nueva_autocorrelacion))
```

```python
maximos = where((abs(nueva_autocorrelacion)>0.00062))[0]

print maximos

valores = -3.0 + maximos*dl

print valores
```
