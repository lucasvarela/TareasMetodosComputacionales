## Convolución Pasabajas

Acá creamos un bump gaussiano de la forma: 
\begin{equation}
e^{-0.05t²}
\end{equation}
para filtrar las frecuencias altas. Se extrapola a dos dimensiones. Luego se saca la transformada de fourier de ambos y se multiplicaron. Se saca la tranformada inversa.

```python
t = np.linspace(-10, 10, 20)
bump = np.exp(-0.05*t**2)
bump = bump / np.trapz(bump) # normalize the integral to 1

# make a 2-D kernel out of it
kernel = bump[:, newaxis] * bump[newaxis, :]
print Im1.shape[:2]

# padded fourier transform, with the same shape as the image
kernel_ft = fftpack.fft2(kernel,shape=Im1.shape[:2], axes=(0, 1))

# convolve

img_ft = fftpack.fft2(Im1, axes=(0, 1))
print shape(kernel_ft)
print shape(img_ft)
img2_ft = kernel_ft[:, :] * img_ft
img2 = fftpack.ifft2(img2_ft, axes=(0, 1)).real
```

Se grafica la imagen original

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(Im1[:,:], cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)
```

Se grafica la imagen con el filtro

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(img2[:,:], cmap = cm.Greys_r)
plt.title('pasa baja', fontsize=12)
```

Como se ve en la imagen, al quitar las frecuencias altas los bordes se difuminan.

## Convolución Pasaaltas

Acá creamos un bump gaussiano de la forma: 
\begin{equation}
1-e^{-0.05t²}
\end{equation}
para filtrar las frecuencias bajas. Se extrapola a dos dimensiones. Luego se saca la transformada de fourier de ambos y se multiplicaron. Se saca la tranformada inversa.

```python
t = np.linspace(-10, 10, 20)
bump = 1-np.exp(-0.05*t**2)
bump = bump / np.trapz(bump) # normalize the integral to 1

# make a 2-D kernel out of it
kernel = bump[:, newaxis] * bump[newaxis, :]
print Im1.shape[:2]

# padded fourier transform, with the same shape as the image
kernel_ft = fftpack.fft2(kernel,shape=Im1.shape[:2], axes=(0, 1))

# convolve

img_ft = fftpack.fft2(Im1, axes=(0, 1))
print shape(kernel_ft)
print shape(img_ft)
img2_ft = kernel_ft[:, :] * img_ft
img2 = fftpack.ifft2(img2_ft, axes=(0, 1)).real
```

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(Im1[:,:], cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)
```

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(img2[:,:], cmap = cm.Greys_r)
plt.title('pasa altas', fontsize=12)
```

Como se ve en la imagen al quitar las frecuencias bajas, las irregularidades se amplifican. Quitar las frecuencias bajas implica que las partes de la señal que varian poco a lo largo de la imagen.
