```python
def leer_tiff(filename):
    
    imagen = imread(filename)
    
    return imagen

def normalizar_tiff(imagen):
    
    maximo = amax(imagen)
    imagen = imagen/float(maximo)
    
    return imagen

def likelihood(imagen_original, imagen_modelo):
    
    chi_squared = (1.0/2.0)*sum((imagen_original-imagen_modelo)**2)
    
    return -chi_squared

def ecuacion(thetas, r_0, r_1, phi, m):
    
    return r_0 + r_1*sin( m*thetas + phi )

def modelo(imagen, thetas, r_0, r_1, phi, m, c_x, c_y):
    
    imagen_modelo = zeros(shape(imagen))

    r = ecuacion(thetas, r_0, r_1, phi, m)

    x = r*cos(thetas) + c_x
    y = r*sin(thetas) + c_y
 
    for i in range(len(x)):
        
        ind_x = int(x[i])
        ind_y = int(y[i])
        
        imagen_modelo[(ind_x,ind_y)] = 1.0
    
    return imagen_modelo
```

# Se prueba con parametros iniciales que a ojo dan un buen punto de inicio para no desperdiciar iteraciones comenzando en algun punto aleatorio.

```python

fig = figure(figsize=(7,7))

imshow( modelo(imagen, thetas, 105, 10, 2.5, 1, 150, 140) , cmap = 'bone')
```

```python
r_0_walk = empty((0))
r_1_walk = empty((0))
phi_walk = empty((0))
m_walk = empty((0)) 
c_x_walk = empty((0)) 
c_y_walk = empty((0)) 

l_walk = empty((0))

r_0_walk = append(r_0_walk, 105)
r_1_walk = append(r_1_walk, 10)
phi_walk = append(phi_walk, 2.5)
m_walk = append(m_walk, 1)
c_x_walk = append(c_x_walk, 150)
c_y_walk = append(c_y_walk, 140)

imagen_modelo_init = modelo(imagen, thetas, r_0_walk[0], r_1_walk[0], phi_walk[0], m_walk[0], c_x_walk[0], c_y_walk[0])
l_walk = append(l_walk, likelihood(imagen, imagen_modelo_init))

print r_0_walk
print r_1_walk
print phi_walk
print m_walk
print c_x_walk
print c_y_walk

print l_walk
```

```python
n_iterations = 10001

for i in range(n_iterations):
    
    r_0_prime = random.normal(r_0_walk[i], 0.1) 
    r_1_prime = random.normal(r_1_walk[i], 0.1) 
    phi_prime = random.normal(phi_walk[i], 0.1) 
    m_prime = random.normal(m_walk[i], 0.1) 
    c_x_prime = random.normal(c_x_walk[i], 0.1) 
    c_y_prime = random.normal(c_y_walk[i], 0.1) 

    imagen_modelo_init = modelo(imagen, thetas, r_0_walk[i], r_1_walk[i], phi_walk[i], m_walk[i], c_x_walk[i], c_y_walk[i])
    imagen_modelo_prime = modelo(imagen, thetas, r_0_prime, r_1_prime, phi_prime, m_prime, c_x_prime, c_y_prime)
    
    l_prime = likelihood(imagen, imagen_modelo_prime)
    l_init = likelihood(imagen, imagen_modelo_init)
    
    alpha = l_prime-l_init
    
    if(alpha >= 0.0):
        r_0_walk  = append(r_0_walk, r_0_prime)
        r_1_walk  = append(r_1_walk, r_1_prime)
        phi_walk  = append(phi_walk, phi_prime)
        m_walk  = append(m_walk, m_prime)
        c_x_walk  = append(c_x_walk, c_x_prime)
        c_y_walk  = append(c_y_walk, c_y_prime)
        
        l_walk = append(l_walk, l_prime)
    
    else:
        beta = random.random()
        
        if(beta <= exp(alpha)):
            r_0_walk  = append(r_0_walk, r_0_prime)
            r_1_walk  = append(r_1_walk, r_1_prime)
            phi_walk  = append(phi_walk, phi_prime)
            m_walk  = append(m_walk, m_prime)
            c_x_walk  = append(c_x_walk, c_x_prime)
            c_y_walk  = append(c_y_walk, c_y_prime)
        
            l_walk = append(l_walk, l_prime)
            
        else:
            r_0_walk  = append(r_0_walk, r_0_walk[i])
            r_1_walk  = append(r_1_walk, r_1_walk[i])
            phi_walk  = append(phi_walk, phi_walk[i])
            m_walk  = append(m_walk, m_walk[i])
            c_x_walk  = append(c_x_walk, c_x_walk[i])
            c_y_walk  = append(c_y_walk, c_y_walk[i])
        
            l_walk = append(l_walk, l_init)
            
    if(i%1000 == 0):
        print "Avance:", i*100/(n_iterations-1),"%"
```

```python
fig, ax = plt.subplots(2, 3, figsize=(17, 10))

ax[0,0].scatter(r_0_walk, -l_walk, alpha=0.2)
ax[0,0].set_xlabel("$r_0$")
ax[0,0].set_ylabel("$-\mathcal{L}$")

ax[0,1].scatter(r_1_walk, -l_walk, alpha=0.2)
ax[0,1].set_xlabel("$r_1$")
ax[0,1].set_ylabel("$-\mathcal{L}$")

ax[0,2].scatter(phi_walk, -l_walk, alpha=0.2)
ax[0,2].set_xlabel("$\phi$")
ax[0,2].set_ylabel("$-\mathcal{L}$")

ax[1,0].scatter(m_walk, -l_walk, alpha=0.2)
ax[1,0].set_xlabel("$m$")
ax[1,0].set_ylabel("$-\mathcal{L}$")

ax[1,1].scatter(c_x_walk, -l_walk, alpha=0.2)
ax[1,1].set_xlabel("$c_x$")
ax[1,1].set_ylabel("$-\mathcal{L}$")

ax[1,2].scatter(c_y_walk, -l_walk, alpha=0.2)
ax[1,2].set_xlabel("$c_y$")
ax[1,2].set_ylabel("$-\mathcal{L}$")
```
