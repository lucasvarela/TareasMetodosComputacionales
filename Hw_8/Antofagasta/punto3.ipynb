```python
archivo=np.loadtxt("antofagasta.dat")
# MAyoria de pasos y formulas se obtuvieron del libro Structural Geology Algorithms, Vectors and Tensors. De Richard Allmendinger
```

```python
#Llamando las filas
longitud=archivo[:,0]
latitud=archivo[:,1]
ulon=archivo[:,2]
errorlon=archivo[:,3]
ulat=archivo[:,4]
errorlat=archivo[:,5]
filas=len(archivo)
pi=3.14159
```

```python
#http://www.gabrielortiz.com/index.asp?Info=058a                Todas las formulas y datos se obtuvieron de aca
#Convirtiendo coordenadas geograficas a UTM
#In UTM coordinates, Chile is in Zone 19S

ejemayor=6378388.0
ejemenor=6356911.946130
excentricidad=(((ejemayor**2)-(ejemenor**2))**(0.5))/ejemayor
excentricidad2=(((ejemayor**2)-(ejemenor**2))**(0.5))/ejemenor
exe=excentricidad2**2
rpc=(ejemayor**2)/ejemenor
aplana=(ejemayor-ejemenor)/ejemayor
husototal=(-70/6+31) #Promedio de los datos aproximado de la longitud es de 70, todos caen en el huso 19S
husocentral=husototal*6-183
alfa1=(3.0/4.0)*exe
beta1=(5.0/3.0)*alfa1**2
gamma1=(35.0/27.0)*alfa1**3
newlat=zeros(filas)
newlon=zeros(filas)
deltahuso=zeros(filas)
paraA=zeros(filas)
paraE=zeros(filas)
paraM=zeros(filas)
paraV=zeros(filas)
paraC=zeros(filas)
paraA1=zeros(filas)
paraA2=zeros(filas)
paraJ2=zeros(filas)
paraJ4=zeros(filas)
paraJ6=zeros(filas)
paraB=zeros(filas)
longitudfinal=zeros(filas)
latitudfinal=zeros(filas)
```

```python
#Transformado coordenadas a radianes
for i in range (filas):
    newlat[i]=latitud[i]*pi/180

    newlon[i]=longitud[i]*pi/180
    
```

```python
for i in range (filas):
    deltahuso[i]=newlon[i]-husocentral*pi/180

for i in range (filas):
    paraA[i]=cos(newlat[i])*sin(deltahuso[i])

for i in range (filas):
    paraE[i]=(0.5)*np.log((1+paraA[i])/(1-paraA[i]))
    
for i in range (filas):
    paraM[i]=(np.arctan((tan(newlat[i]))/(cos(deltahuso[i]))))-newlat[i]
    
for i in range (filas):
    paraV[i]=(0.9996)*(rpc/((1+(exe)*(cos(newlat[i]))**2)**(0.5)))
    
for i in range (filas):
    paraC[i]=((exe)/2)*(paraE[i]**2)*((cos(newlat[i]))**2)
    
for i in range (filas):
    paraA1[i]=sin(2*newlat[i])
    
for i in range (filas):
    paraA2[i]=paraA1[i]*((cos(newlat[i]))**2)
    
for i in range (filas):
    paraJ2[i]=(newlat[i])+(paraA1[i]/2)
    
for i in range (filas):
    paraJ4[i]=(3*paraJ2[i]+paraA2[i])/4
    
for i in range (filas):
    paraJ6[i]=(5*paraJ4[i]+(paraA2[i])*((cos(newlat[i]))**2))
    
for i in range (filas):
    paraB[i]=(0.9996*rpc)*(newlat[i]-alfa1*paraJ2[i]+beta1*paraJ4[i]-gamma1*paraJ6[i])
```

```python
#COordenadas UTM finales

for i in range (filas):
    longitudfinal[i]=(paraE[i]*paraV[i]*(1+paraC[i]/3.0)+500000)
    
    latitudfinal[i]=(paraM[i]*paraV[i]*(1+paraC[i])+paraB[i])+10000000             #Se suma 10.000.000 si se esta debajo del Ecuador
    
```

```python
#Posicion de estaciones iniciales en UTM
A=plt.scatter(longitudfinal[:],latitudfinal[:],color='blue')
plt.title("Coordenadas en UTM")
plt.xlabel("longitud")
plt.ylabel("latitud")
```

```python
magnitud=zeros(filas)
angulo=zeros(filas)
magnituddef=zeros(filas)
magnitudpromedio=0.0
angulopromedio=0.0
cos1=0.0
cos2=0.0
for i in range (filas):
    magnitud[i]=(ulat[i]**2+ulon[i]**2)**(0.5)
    angulo[i]=arctan(ulat[i]/ulon[i])
    magnitudpromedio=magnitudpromedio+magnitud[i]/filas
    angulopromedio=angulopromedio+angulo[i]/filas
    cos1=cos1+cos(angulo[i])
    cos2=cos2+cos(90-angulo[i])
    
```

```python
n=((cos1**2+cos2**2)**(0.5))
r=n/filas
#Entre mas cercano sea el r a 1, mejor concentrados estan los datos
r,n, magnitudpromedio, angulopromedio
```

### #Transformando los ejes

```python
cosdirector=np.array([[cos1/n,cos2/n],[-cos2/n,cos1/n]])  #matriz de deformacion
#http://ocwus.us.es/mecanica-de-medios-continuos-y-teoria-de-estructuras/calculo-de-estructuras-1/apartados/apartado6_2.html/skinless_view
```

#### En las siguientes graficas, se puede deducir que la deformacion aumenta entre mas al Occidente se vaya, indicando que la falla tiene rumbo este-oeste, y el bloque colgante se movio hacia el Occidente. El epicentro del terremoto se  encontraria hacia el OCcidente, donde las magnitudes son mayores. No se observa un patron ploteando la deformacion contra la latitud.

```python
#Transformando Todas las cosas

latfinal=zeros(filas)
lonfinal=zeros(filas)
errorlatfinal=zeros(filas)
errorlonfinal=zeros(filas)

for i in range (filas):
    lonfinal[i]=(cos1/n)*longitudfinal[i]-(cos2/n)*latitudfinal[i]
    latfinal[i]=(cos2/n)*longitudfinal[i]+(cos1/n)*latitudfinal[i]
    errorlonfinal[i]=(cos1/n)*errorlon[i]-(cos2/n)*errorlat[i]
    errorlatfinal[i]=(cos2/n)*errorlon[i]+(cos1/n)*errorlat[i]
```

#### Se puede observar un patron graficando la magnitud de la deformacion vs la longitud, se observa que en las longitudes hacia el Oeste hay mayor deformaciòn, indicando que el epicentro del terremoto se encuentra hacia el Oeste. Desafortunadamente, el  patron no es suficiente como para hacer un fit total, es mejor dividirla por partes.

#### Tensor de deformacion final, (1/2)(eij+eji)

```python
eij=np.array([[dudx,dudy],[dvdx,dvdy]])
eji=np.array([[dudx,dvdx],[dudy,dvdy]])
```

```python
tensor=(0.5)*(eij+eji) #tensor final
```
